```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Ładowanie danych z pliku z predykcjami
try:
    results = pd.read_csv('C:\\Users\\mpiesio\\Desktop\\KODILLA\\wizualizacja_3\\spam_predictions.csv')
    y_test = results['True_Label']
    y_pred = results['Predicted_Label']
except FileNotFoundError:
    print("Błąd: Plik spam_predictions.csv nie istnieje. Upewnij się, że plik został zapisany.")
    exit()

# Obliczanie metryk klasyfikacji
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='binary')
recall = recall_score(y_test, y_pred, average='binary')
f1 = f1_score(y_test, y_pred, average='binary')

# Wyświetlanie metryk
print("\nMetryki klasyfikacji na zbiorze testowym:")
print(f"Dokładność (Accuracy): {accuracy:.4f}")
print(f"Precyzja (Precision): {precision:.4f}")
print(f"Czułość (Recall): {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# Obliczanie macierzy omyłek
conf_matrix = confusion_matrix(y_test, y_pred)

# Wyświetlanie macierzy omyłek
print("\nMacierz omyłek:")
print(conf_matrix)

# Wizualizacja macierzy omyłek
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Nie-Spam (0)', 'Spam (1)'], 
            yticklabels=['Nie-Spam (0)', 'Spam (1)'])
plt.xlabel('Predykcja')
plt.ylabel('Rzeczywista klasa')
plt.title('Macierz omyłek dla modelu klasyfikacji spamu')
plt.tight_layout()
plt.show()

# Zapis macierzy omyłek do pliku (opcjonalne)
np.savetxt('C:\\Users\\mpiesio\\Desktop\\KODILLA\\wizualizacja_3\\confusion_matrix.csv', 
           conf_matrix, delimiter=',', fmt='%d')
print("\nMacierz omyłek zapisano do 'confusion_matrix.csv'.")
```
